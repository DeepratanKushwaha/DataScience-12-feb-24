    uk ---> london
    
    aus  --> australia

```python
def func(name):
    if name.lower() == 'india':
        return 'Bharat'
    elif name.lower() == 'aus':
        return 'Australia'
    elif name.lower() == 'usa':
        return 'america'
    elif name.lower() == 'uk':
        return 'london'
    else:
        return 'Invalid/Nan'
```

```python
df
```

```python
df['Country'].apply(func)
```

```python
df['Country'] = df['Country'].apply(func)
```

```python
df
```
