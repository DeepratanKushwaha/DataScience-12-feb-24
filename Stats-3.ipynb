## Shape of Data


            1. Normal distribution
            
            2. Uniform Distribution

```python
data = list(np.random.normal(100, 20, 1000))
print(data)
```

```python
plt.hist(data,bins=30, ec='k')
plt.show()
```

```python
mean = np.mean(data)
```

```python
std = np.std(data)
```

```python
plt.figure(figsize=(16,9), dpi=200)
plt.hist(data,bins=30, ec='k')
plt.title('Normal Distribution', size=20)
plt.plot([mean, mean], [0, 100], label='Mean')
plt.plot([mean-std, mean-std], [0,100], label='STD')
plt.plot([mean+std, mean+std], [0,100], label='STD')
plt.xlabel('Data', size=20)
plt.ylabel('Frequency', size=20)
plt.legend()
plt.show()
```

__In Normal distribution  std cover majority of data points / observation__

```python
arr = list(np.random.uniform(100, 250, 1000))
```

```python
print(arr)
```

```python
plt.figure(figsize=(16,9), dpi=200)
plt.hist(arr, bins=30, ec='r')
mean = np.mean(arr)
dev = np.std(arr)
plt.title('Uniform Distribution', size=20)
plt.xlabel('Data', size=20)
plt.ylabel('Frequency', size=20)
plt.plot([mean,mean], [0, 60], lw=2)
plt.plot([mean-std, mean-std], [0, 60], lw=2)
plt.plot([mean+std, mean+std], [0, 60], lw=2)
plt.show()
```

    In above histogram data is equally distributed among all categories
    So if we use std to cover all data points it is not possible because sometimes data can be skewed.
    that's where we use quantilies.

    mean > median     +ve deviation
    
    mean < median    - ve deviation
