```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style
import seaborn as sns
```

```python
x = np.arange(1, 30, 2)
y = x ** 2
```

    mfc  --> marker face color

    mec  ---> marker edge color

    ms  ----> marker size

    lw ----> line width

```python
plt.figure(figsize=(14, 8), dpi=300, facecolor='k')
plt.plot(x, '--r*', label='X', )
plt.plot(y, '--bh', label='X^2')
plt.xlabel('Index', size=20, color='w')
plt.ylabel('Data', size=20, color='w')
plt.tick_params(color='w', labelcolor='w')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14, 8), dpi=300, facecolor='k')
plt.plot(x, '--r*', mfc='b', mec='y', ms=10, label='X', )
plt.plot(y, '--bh', mfc='r', mec='k', label='X^2')
plt.xlabel('Index', size=20, color='w')
plt.ylabel('Data', size=20, color='w')
plt.tick_params(color='w', labelcolor='w')
plt.legend()
plt.show()
```

```python
taxis = sns.load_dataset('taxis')
taxis
```

```python
taxis.isna().sum()
```

```python
taxis.dropna(inplace=True)
```

```python
taxis
```

```python
plt.figure(figsize=(16, 9), dpi=300)
plt.scatter(taxis['distance'], taxis['fare'], color='r')
plt.show()
```

```python
taxis['payment'].value_counts()
```

```python
color = taxis['color'].value_counts()
color
```

```python
colors = taxis['color'].map({'yellow':'y', 'green':'g'})
colors
```

```python
size = taxis['payment'].map({'credit card':10, 'cash':50})
size
```

```python
plt.figure(figsize=(16, 9), dpi=300)
plt.scatter(taxis['distance'], taxis['fare'], color=colors, s=size)
plt.plot([],[], 'yo', label='yellow taxis')
plt.plot([],[], 'go', label='green taxis')
plt.plot([],[],'o', label='size: 10 credit')
plt.plot([],[], 'o', label='size: 50 cash')

plt.legend()
plt.show()
```

```python
taxis.groupby('pickup_zone')[['passengers']].sum()
```

```python
taxis.groupby('pickup_zone')[['passengers']].sum().sort_values(by='passengers', ascending=False)
```

```python
taxis.groupby('pickup_zone')[['passengers']].sum().sort_values(by='passengers', ascending=False).head(20)
```

```python
index = taxis.groupby('pickup_zone')[['passengers']].sum().sort_values(by='passengers', ascending=False).head(20).index
values = taxis.groupby('pickup_zone')[['passengers']].sum().sort_values(by='passengers', ascending=False).head(20).values
```

```python
index
```

```python
flat_values = values.flatten()
```

```python
plt.figure(figsize=(16, 9), dpi=300, facecolor='k')
plt.bar(index, flat_values, color=['r','b','g',], alpha=0.6)
plt.xlabel('Pickup_Zone', size=20, color='y')
plt.ylabel('Count', size=20, color='y')
plt.title('Total Number of Passengers in Top 20 pickup zone', size=20, color='y')
plt.tick_params(color='w', labelcolor='w', rotation=90, labelsize=20)
plt.show()
```
