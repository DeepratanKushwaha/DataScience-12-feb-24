```python
days = ['sun', 'mon', 'tues', 'wednes', 'thrus', 'fri', 'sat']
```

```python
jaipur_temp = [23.4, 26.5, 28.7, 34, 30, 28, 27]
```

```python
plt.plot(days, jaipur_temp)
```

```python
delhi_temp = [33, 34.4, 45, 40, 36.5, 29.9, 32]
```

```python
plt.figure(figsize=(16, 9))
plt.plot(days, jaipur_temp, days, delhi_temp)
plt.title('Jaipur vs Delhi Temp', size=20)
plt.ylabel('Temp', size=20)
plt.xlabel('Days', size=20)
plt.show()
```

```python
plt.figure(figsize=(16, 9))
plt.plot(days, jaipur_temp, 'ro--', lw=2, label='Jaipur_Temp')
plt.plot(days, delhi_temp, 'y+:', lw=3, label='Delhi_Temp')
plt.title('Jaipur vs Delhi Temp', size=20)
plt.ylabel('Temp', size=20)
plt.xlabel('Days', size=20)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16, 9))
plt.plot(days, jaipur_temp, 'ro--', lw=2, label='Jaipur_Temp')
plt.plot(days, delhi_temp, 'y+:', lw=3, label='Delhi_Temp')
plt.title('Jaipur vs Delhi Temp', size=20)
plt.ylabel('Temp', size=20)
plt.xlabel('Days', size=20)
plt.xticks(fontsize=20, rotation=90)
plt.yticks(fontsize=20)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16, 9))
plt.plot(days, jaipur_temp, 'ro--', lw=2, label='Jaipur_Temp')
plt.plot(days, delhi_temp, 'y+:', lw=3, label='Delhi_Temp')
plt.title('Jaipur vs Delhi Temp', size=20)
plt.ylabel('Temp', size=20)
plt.xlabel('Days', size=20)
plt.tick_params(size=20)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16, 9))
plt.plot(days, jaipur_temp, 'ro--', lw=2, label='Jaipur_Temp')
plt.plot(days, delhi_temp, 'y+:', lw=3, label='Delhi_Temp')
plt.title('Jaipur vs Delhi Temp', size=20)
plt.ylabel('Temp', size=20)
plt.xlabel('Days', size=20)
plt.tick_params(labelsize=20)
plt.legend()
plt.show()
```

```python
data = pd.read_csv('netflix-rotten-tomatoes-metacritic-imdb.csv')
data
```

```python
data['Series or Movie'].value_counts()
```

```python
data['Series or Movie'].value_counts().plot(kind='bar')
```

```python
data['Series or Movie'].value_counts()
```

```python
index = data['Series or Movie'].value_counts().index
index
```

```python
values = data['Series or Movie'].value_counts().values
values
```

```python
plt.figure(figsize=(16, 9), dpi=300)
plt.bar(index, values, color=['red', 'yellow'])
plt.xlabel('Categories', size=20)
plt.ylabel('Count', size=20)
plt.tick_params(labelsize=15)
plt.show()
```
