## Pie chart

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.style as style
```

```python
data = [20, 30, 34, 45, 23, 78]
subject = ['ML', 'python', 'Data science', 'full stack', 'devops', 'linux']
```

```python
plt.pie(data)
plt.show()
```

```python
plt.pie(data, labels=subject)
plt.show()
```

```python
plt.pie(data, labels=subject, explode=[0,0,0,0,0.2,0])
plt.show()
```

```python
plt.pie(data, labels=subject, explode=[0,0.1,0,0,0.2,0.5])
plt.show()
```

```python
plt.pie(data, labels=subject, explode=[0,0.1,0,0,0.2,0.5], shadow=True)
plt.show()
```

```python
plt.pie(data, labels=subject, explode=[0,0.1,0,0,0.2,0.5], shadow=True, autopct='%0.2f%%')
plt.show()
```

```python
plt.pie(data, labels=subject, explode=[0,0.1,0,0,0.2,0.5], shadow=True, autopct='%0.4f%%')
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.pie(data, labels=subject, explode=[0,0.1,0,0,0.2,0.5], shadow=True, autopct='%0.4f%%')
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.pie(data, labels=subject, explode=[0,0.1,0,0,0.2,0.5], shadow=True, autopct='%0.4f%%', counterclock=False)
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.pie(data, labels=subject, explode=[0,0.1,0,0,0.2,0.5], shadow=True, autopct='%0.4f%%', counterclock=False)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.pie(data, labels=subject, explode=[0,0.1,0,0,0.2,0.5], shadow=True, autopct='%0.4f%%', counterclock=False)
plt.legend()
plt.show()
```
