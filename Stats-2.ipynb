## Measures of Spread

    1. Deviation
    2. Standard Deviation
    3. Variance

## deviation
                    
                    dev = mean - data

```python
arr = list(np.random.randint(20, 35, 100))
```

```python
print(arr)
```

```python
mean = np.mean(arr)
mean
```

```python
dev = [(mean-i) for i in arr]
print(dev)
```

```python
v = sum(dev) / len(dev)
v
```

        This tells us that the points are not getting deviate from mean 
    

```python
v +5
```

```python
dev = [abs(mean-i) for i in arr]
print(dev)
```

```python
np.mean(dev)
```

```python
np.std(arr)
```

### Drawback of deviation
        Deviation only covers the nearby values. So the large values
        get neglected
        
        So we will overcome this drawback using variance

### Variance

$$ variance = \frac{\sum\limits_{i=1}^N (x_i - \bar x)^2} {N} $$

```python
print(arr)
```

```python
var = [(mean-i)**2 for i in arr]
print(var)
```

```python
sum(var) / len(var)
```

```python
arr = list(np.random.randint(20, 35, 100))
print(arr)
```

```python
mean = np.mean(arr)
mean
```

```python
std = np.std(arr)
std
```

```python
plt.figure(figsize=(16, 9), dpi=200)
plt.plot(arr, 'or', label='Ages')
plt.plot([0,100], [mean, mean], label="Mean")
plt.plot([0,100], [mean+std, mean+std], label='STD')
plt.plot([0,100], [mean-std, mean-std], label='STD')
plt.xlabel('Index', size=20)
plt.ylabel('Ages', size=20)
plt.tick_params(labelsize=15)
plt.legend()
plt.show()
```

```python
var = np.var(arr)
var
```

```python
plt.figure(figsize=(16, 9), dpi=200)
plt.plot(arr, 'or', label='Ages')
plt.plot([0,100], [mean, mean], label="Mean")
plt.plot([0,100], [mean+std, mean+std], label='STD')
plt.plot([0,100], [mean-std, mean-std], label='STD')
plt.plot([0,100], [mean+var, mean+var], label='VAR')
plt.plot([0,100], [mean-var, mean-var], label='VAR')
plt.xlabel('Index', size=20)
plt.ylabel('Ages', size=20)
plt.tick_params(labelsize=15)
plt.legend()
plt.show()
```

#### In Variance the outliers are also covered. Which will be solved by STD

```python
plt.figure(figsize=(16, 9), dpi=200)
plt.plot(arr, 'or', label='Ages')
plt.plot([0,100], [mean, mean], label="Mean")
plt.plot([0,100], [mean+std, mean+std], label='STD')
plt.plot([0,100], [mean-std, mean-std], label='STD')
plt.plot([0,100], [mean+2*std, mean+2*std], label='2*STD')
plt.plot([0,100], [mean-2*std, mean-2*std], label='2*STD')
plt.xlabel('Index', size=20)
plt.ylabel('Ages', size=20)
plt.tick_params(labelsize=15)
plt.legend()
plt.show()
```
