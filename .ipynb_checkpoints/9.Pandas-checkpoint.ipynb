```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame({
    'country':np.random.choice(['India', 'China', 'Usa', 'Aus'], 20, True),
    'gender':np.random.choice(['Male', 'Female'], 20, True),
    'salary':np.random.randint(40, 90, 20),
    'height':np.random.randint(120, 190, 20),
    'weight':np.random.randint(50, 100, 20)
}, index=[chr(65+i) for i in range(20)])
```

```python
df2.unstack().unstack()
```

       to change the view of data we can perform stacking and unstacking
       
       stack  -->  long to wide 
    
        unstack ---> wide to long
