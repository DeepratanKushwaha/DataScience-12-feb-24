```python
import requests

fsq_id = "4bc8f091762beee173b93d38"

url = f"https://api.foursquare.com/v3/places/{fsq_id}/photos"
api_key = "fsq3/qWKdb9qPqlEThP0P7+f2kX52mr7qCPxMoBZAg3Ba/U="

params = {
    "limit":50,
    "sort":"newest",
    "classification":"indoor"
}

headers = {"accept": "application/json",
          "Authorization": api_key
          }

resp = requests.get(url, params=params, headers=headers)
resp
```

```python
resp.content
```

```python
data = resp.json()
data
```

```python
data[0]
```

```python
prefix = data[0]['prefix']
suffix = data[0]['suffix']
```

```python
print(prefix+'original'+suffix)
```

```python
urls = []
for i in data:
    prefix = i['prefix']
    suffix = i['suffix']
    urls.append((prefix+'original'+suffix))
```

```python
print(urls)
```

```python
urls[0]
```

```python
resp = requests.get(urls[0])
resp
```

```python
resp.content
```

```python
with open('img.jpeg', 'wb') as f:
    f.write(resp.content)
    
```
