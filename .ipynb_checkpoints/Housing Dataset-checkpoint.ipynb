```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
house = pd.read_csv('housing.csv')
house
```

```python
house['ocean_proximity'].value_counts()
```

### median house value according to ocean proximity

```python
house.groupby(['ocean_proximity'])[['median_house_value']].median()
```

```python
house.groupby(['ocean_proximity'])[['median_house_value']].median().sort_values(by='median_house_value',ascending=False)
```

### total population according to ocean proximity

```python
house.head()
```

```python
house.groupby('ocean_proximity')[['population']].sum().sort_values(by='population', ascending=False)
```

```python
house.describe()
```

```python
new_houses = house[house['housing_median_age'] < 15]
new_houses
```

```python
old_houses = house[house['housing_median_age'] > 45]
old_houses
```

## percentage of old houses and new houses on near bay ocean proximity

```python
new_houses['ocean_proximity'].value_counts()
```

```python
new_houses['ocean_proximity'].value_counts() / new_houses.shape[0] * 100
```

```python
old_houses['ocean_proximity'].value_counts() / old_houses.shape[0] * 100
```
