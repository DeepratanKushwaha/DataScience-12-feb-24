## histogram

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ml_student_age = np.random.randint(20, 35, 100)
py_student_age = np.random.randint(15, 30, 100)
```

```python
ml_student_age
```

```python
py_student_age
```

```python
import matplotlib.style as style
```

```python
plt.figure(figsize=(14,8), dpi=200, facecolor='pink')
style.use('ggplot')
plt.hist(ml_student_age, bins=10, ec='k', color='y')
plt.xlabel('ML Student age', size=20)
plt.ylabel('Frequency', size=20)
plt.title('ML student age distribution', size=20)
plt.tick_params(labelsize=15)
plt.show()
```

```python
plt.figure(figsize=(14,8), dpi=200, facecolor='pink')
style.use('ggplot')
plt.hist(ml_student_age, bins=10, ec='k', color='y', histtype='step')
plt.xlabel('ML Student age', size=20)
plt.ylabel('Frequency', size=20)
plt.title('ML student age distribution', size=20)
plt.tick_params(labelsize=15)
plt.show()
```

```python
plt.figure(figsize=(14,8), dpi=200, facecolor='pink')
style.use('ggplot')
plt.hist(ml_student_age, bins=10, ec='k', color='y', histtype='stepfilled')
plt.xlabel('ML Student age', size=20)
plt.ylabel('Frequency', size=20)
plt.title('ML student age distribution', size=20)
plt.tick_params(labelsize=15)
plt.show()
```

```python
plt.figure(figsize=(14,8), dpi=200, facecolor='pink')
style.use('ggplot')
plt.hist(ml_student_age, bins=10, ec='k', color='y', histtype='barstacked')
plt.xlabel('ML Student age', size=20)
plt.ylabel('Frequency', size=20)
plt.title('ML student age distribution', size=20)
plt.tick_params(labelsize=15)
plt.show()
```

```python
plt.figure(figsize=(14,8), dpi=200, facecolor='pink')
style.use('ggplot')
plt.hist(ml_student_age, bins=10, ec='k', color='y')
plt.xlabel('ML Student age', size=20)
plt.ylabel('Frequency', size=20)
plt.title('ML student age distribution', size=20)
plt.tick_params(labelsize=15)
plt.show()
```

```python
plt.figure(figsize=(14,8), dpi=200, facecolor='pink')
style.use('ggplot')
plt.hist(ml_student_age, bins=10, ec='k', color='y', align='left')
plt.xlabel('ML Student age', size=20)
plt.ylabel('Frequency', size=20)
plt.title('ML student age distribution', size=20)
plt.tick_params(labelsize=15)
plt.show()
```

```python
plt.figure(figsize=(14,8), dpi=200, facecolor='pink')
style.use('ggplot')
plt.hist(ml_student_age, bins=10, ec='k', color='y', align='right')
plt.xlabel('ML Student age', size=20)
plt.ylabel('Frequency', size=20)
plt.title('ML student age distribution', size=20)
plt.tick_params(labelsize=15)
plt.show()
```

```python
plt.figure(figsize=(14,8), dpi=200, facecolor='pink')
style.use('ggplot')
plt.hist(ml_student_age, bins=10, ec='k', color='y', align='right')
plt.hist(py_student_age, bins=10, ec='k', align='left')
plt.xlabel('ML Student age', size=20)
plt.ylabel('Frequency', size=20)
plt.title('ML student age distribution', size=20)
plt.tick_params(labelsize=15)
plt.show()
```

```python
plt.figure(figsize=(14,8), dpi=200, facecolor='pink')
style.use('ggplot')
plt.hist(ml_student_age, bins=10, ec='k', color='y',histtype="barstacked", label='ML')
plt.hist(py_student_age, bins=10, ec='k', label='Python' )
plt.xlabel('ML Student age', size=20)
plt.ylabel('Frequency', size=20)
plt.title('ML student age distribution', size=20)
plt.tick_params(labelsize=15)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14,8), dpi=200, facecolor='pink')
style.use('ggplot')
plt.hist(ml_student_age, bins=10, ec='k', color='y',histtype="barstacked", label='ML')
plt.hist(py_student_age, bins=10, ec='k', label='Python' )
plt.xlabel('ML Student age', size=20)
plt.ylabel('Frequency', size=20)
plt.title('ML student age distribution', size=20)
plt.tick_params(labelsize=15)
plt.legend(loc=2)
plt.show()
```

```python
plt.figure(figsize=(14,8), dpi=200, facecolor='pink')
style.use('ggplot')
plt.hist(ml_student_age, bins=10, ec='k', color='y',histtype="barstacked", label='ML')
plt.hist(py_student_age, bins=10, ec='k', label='Python' )
plt.xlabel('ML Student age', size=20)
plt.ylabel('Frequency', size=20)
plt.title('ML student age distribution', size=20)
plt.tick_params(labelsize=15)
plt.legend(loc=5)
plt.show()
```

```python
plt.figure(figsize=(14,8), dpi=200, facecolor='pink')
style.use('ggplot')
plt.hist([ml_student_age, py_student_age],ec='k', bins=10, histtype="barstacked", label='ML')

plt.xlabel('ML Student age', size=20)
plt.ylabel('Frequency', size=20)
plt.title('ML student age distribution', size=20)
plt.tick_params(labelsize=15)
plt.legend(loc=5)
plt.show()
```

```python
taxis = sns.load_dataset('taxis')
taxis
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
index = tips['sex'].value_counts().index
value = tips['sex'].value_counts().values
```

```python
plt.figure(figsize=(16,9), dpi=300, facecolor='k')
plt.subplot(2,2,1)
plt.scatter(taxis['distance'], taxis['fare'])
plt.xlabel('Distance', color='y')
plt.ylabel('fare', color='y')
plt.title('Distance vs fare', color='y')
plt.tick_params(labelcolor='y')

plt.subplot(2,2,2)
plt.scatter(tips['total_bill'], tips['tip'], c='b')
plt.xlabel('total bill', color='y')
plt.ylabel('tip', color='y')
plt.title('total bill vs fare', color='y')
plt.tick_params(labelcolor='y')

plt.subplot(2,2,3)
plt.bar(index, value, color=['y','g'])
plt.title('Count of Gender', color='y')
plt.xlabel('Gender', color='y')
plt.ylabel('Count', color='y')
plt.tick_params(labelcolor='y')


plt.subplot(2,2,4)
plt.hist(tips['total_bill'], bins=10, ec='y')
plt.xlabel('total bill', color='y')
plt.ylabel('frequency', color='y')
plt.title('Distribution of total bill', color='y')
plt.tick_params(labelcolor='y')
```
