```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tips['time'].value_counts()
```

```python
index = tips['time'].value_counts().index
values = tips['time'].value_counts().values
```

```python
plt.figure(figsize=(16,9), dpi=300)
plt.bar(index, values, color=['r', 'b'])
plt.xlabel('Time', size=15)
plt.ylabel('Count', size=15)
plt.show()
```

```python
plt.figure(figsize=(16, 9), dpi=200)
plt.plot(tips['total_bill'], tips['tip'], 'or')
plt.xlabel('Total Bill', size=20)
plt.ylabel('Tips', size=20)
plt.tick_params(labelsize=15)
plt.show()
```

```python
import matplotlib.style as style
```

```python
style.available
```

```python
style.use('ggplot')
plt.figure(figsize=(16, 9), dpi=200, facecolor='y')
plt.plot(tips['total_bill'], tips['tip'], 'or')
plt.xlabel('Total Bill', size=20, color='k')
plt.ylabel('Tips', size=20, color='k')
plt.tick_params(labelsize=15, labelcolor='k')
plt.show()
```

```python
style.use('seaborn-v0_8-colorblind')
plt.scatter(tips['total_bill'], tips['tip'])
plt.show()
```

```python
style.use('seaborn-v0_8-colorblind')
plt.scatter(tips['total_bill'], tips['tip'], c='y', s=20)
plt.show()
```

```python
jaipur_student = pd.DataFrame({
    'Day':pd.Categorical(['Sun', 'Mon', 'Tue', 'Wed', 'Thrus', 'Fri', 'Sat']),
    'Student':np.random.randint(25,40,7)
})
```

```python
jaipur_student
```

```python
delhi_student = pd.DataFrame({
    'Day':pd.Categorical(['Sun', 'Mon', 'Tue', 'Wed', 'Thrus', 'Fri', 'Sat']),
    'Student':np.random.randint(30,45,7)
})
```

```python
delhi_student
```

```python
pune_student = pd.DataFrame({
    'Day':pd.Categorical(['Sun', 'Mon', 'Tue', 'Wed', 'Thrus', 'Fri', 'Sat']),
    'Student':np.random.randint(15,40,7)
})
```

```python
pune_student
```

```python
plt.plot(jaipur_student.Day, jaipur_student.Student,'--rd', label='jaipur')
plt.plot(delhi_student.Day, delhi_student.Student, label='delhi')
plt.plot(pune_student.Day, pune_student.Student, label='pune')
plt.legend()
plt.show()
```

```python
plt.plot(jaipur_student.Day, jaipur_student.Student,'--r*', label='jaipur')
plt.plot(delhi_student.Day, delhi_student.Student,'--kH', mfc='y', label='delhi')
plt.plot(pune_student.Day, pune_student.Student, label='pune')
plt.legend()
plt.show()
```
