## Quantile

    Arr --> q1, q2, q3
    
        q2 --> median
        q1 --> median of values less than Q2
        q3 --> median of values greater than q2

```python
data = list(np.random.normal(140, 30, 1000))
```

```python
mean = np.mean(data)
median = np.median(data)
std = np.std(data)
var = np.var(data)
quantile = np.quantile(data, [0.25, 0.50, 0.75])

```
